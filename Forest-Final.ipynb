```python
# checking epsilon for time and iterations
epsilon_lst=[0.001, 0.005, 0.01, 0.05, 0.1]
def find_iter_value (P4, R4, epsilon_lst, discount=0.99,  max_iter=1000):
    iter_lst=[]
    value_lst = []
    time_lst= []
    for epsilon in epsilon_lst:
        mdp = mdptoolbox.mdp.ValueIteration(P4, R4, discount, epsilon , max_iter)
        mdp.run()
        value_lst.append((mdp.V[0]))
        iter_lst.append(mdp.iter)
        time_lst.append(mdp.time)
    return value_lst, iter_lst, time_lst
```

```python
value_lst, iter_lst, time_lst = find_iter_value(P4, R4, epsilon_lst, discount=0.99,  max_iter=1000)
```
